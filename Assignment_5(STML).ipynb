```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
import seaborn as sns

# Load the diabetes dataset
url_diabetes = '/content/diabetes_data_upload.csv'
df_diabetes = pd.read_csv(url_diabetes)

# Split the dataset into features and target
features_diabetes = df_diabetes.drop(columns=['class'])
target_diabetes = df_diabetes['class']

# Create a function to visualize the relationship between each feature and the target
def visualize_feature_vs_target(feature):
    plt.figure(figsize=(8, 6))
    plt.scatter(features_diabetes[feature], target_diabetes, alpha=0.5, c='green', edgecolors='black')
    plt.title(f'{feature} vs Diabetes Risk', fontsize=18, fontweight='bold')
    plt.xlabel(feature, fontsize=14)
    plt.ylabel('Diabetes Risk', fontsize=14)
    plt.show()

# Visualize the relationship between the first 4 features and the target
for col in features_diabetes.columns[:4]:
    visualize_feature_vs_target(col)

# Encode categorical columns using LabelEncoder
le = LabelEncoder()
df_diabetes_encoded = df_diabetes.apply(lambda col: le.fit_transform(col) if col.dtype == 'O' else col)

# Normalize the dataset using StandardScaler
scaler = StandardScaler()
normalized_diabetes = scaler.fit_transform(df_diabetes_encoded.drop(columns=['class']))

# Display the first few rows of the encoded dataset
print(df_diabetes_encoded.head())

# Let's calculate the correlation matrix of the encoded dataset
correlation_matrix = df_diabetes_encoded.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, square=True,
            cbar_kws={'shrink': 0.8}, vmin=-1, vmax=1)
plt.title('Correlation Matrix of Diabetes Dataset', fontsize=18, fontweight='bold')
plt.show()
```
