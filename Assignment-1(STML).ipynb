```python
import numpy as np
import matplotlib.pyplot as plt

# Given dataset
X = np.array([2.3, 2.5, 3.6, 2.8, 3.1, 2.9, 3.2, 2.7, 2.8, 3.0])

# Plotting the histogram with density
plt.hist(X, bins=5, density=True, alpha=0.7, color='yellow')
plt.xlabel('Data')
plt.ylabel('Density')
plt.title('Density Histogram with 5 Bins')
plt.show()
```

```python
import numpy as np
from scipy import stats

# Dependent variables (y)
y = np.array([4, 5, 5, 6, 8, 3, 2, 8, 4, 3])

# Independent variables (x)
x = np.array([2, 7, 5, 8, 5, 8, 4, 3, 2, 7])

def describe_data(data):
    mean = np.mean(data)
    median = np.median(data)
    mode_result = stats.mode(data, keepdims=True)
    mode = mode_result.mode[0]
    variance = np.var(data, ddof=1)
    std_dev = np.std(data, ddof=1)
    data_range = np.ptp(data)
    iqr = stats.iqr(data)
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)

    print(f"Mean: {mean}")
    print(f"Median: {median}")
    print(f"Mode: {mode}")
    print(f"Variance: {variance}")
    print(f"Standard Deviation: {std_dev}")
    print(f"Range: {data_range}")
    print(f"Interquartile Range (IQR): {iqr}")
    print(f"Skewness: {skewness}")
    print(f"Kurtosis: {kurtosis}")
    print()

# Descriptive statistics for y
print("Descriptive Statistics for y:")
describe_data(y)

# Descriptive statistics for x
print("Descriptive Statistics for x:")
describe_data(x)
```

```python
# We will use the same function describe_data to find all statistical measures

# Since no new data is provided, using x and y from Question 2 as features
print("Descriptive Statistics for Feature x:")
describe_data(x)

print("Descriptive Statistics for Feature y:")
describe_data(y)
```
